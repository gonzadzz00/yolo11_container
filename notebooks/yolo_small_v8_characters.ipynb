```python
from ultralytics import YOLO
```

```python
# Cargar el modelo YOLOv8 small preentrenado
model = YOLO("yolov8s.pt")
```

```python

# entrenar con dataset personalizado
model.train(
    data="/home/gonzadzz/GitHub/yolo11_container/Container OCR.v1-ocr-characters-for-shipping-containers.yolov8/data.yaml",
    epochs=300,
    patience=50,          # early stopping
    imgsz=640,
    batch=16,
    device=0,             # GPU 0
    workers=8,
    cache=True,
    lr0=0.005,            # learning rate inicial
    optimizer="SGD",      # m√°s estable que Adam para OCR
    mosaic=0.1,           # bajo para no deformar caracteres
    mixup=0.0,            # desactivar mixup
    hsv_h=0.015,          # aumentos ligeros de color
    hsv_s=0.7,
    hsv_v=0.4,
    project="/home/gonzadzz/GitHub/yolo11_container/YOLO_Characters",          # carpeta ra√≠z
    name="Character_YOLO_container",   # nombre del experimento
    exist_ok=False                      # evita sobrescribir entrenamientos previos
)

```
