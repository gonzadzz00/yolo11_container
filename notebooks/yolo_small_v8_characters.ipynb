```python
from ultralytics import YOLO
```

```python
# Cargar el modelo YOLOv8 small preentrenado
model = YOLO("yolov8s.pt")
```

```python
# entrenar con tu dataset de Roboflow
model.train(
    data="/home/gonzadzz/GitHub/yolo11_container/container-ocr.v11i.yolov11/data.yaml",
    epochs=300,
    patience=50,       # early stopping
    imgsz=640,
    batch=16,
    device=0,          # GPU 0
    workers=8,
    cache=True,
    lr0=0.005,         # learning rate inicial
    optimizer="SGD",   # mÃ¡s estable que Adam para OCR
    mosaic=0.1,        # bajo para no deformar caracteres
    mixup=0.0,         # desactivar mixup
    hsv_h=0.015,       # aumentos ligeros de color
    hsv_s=0.7,
    hsv_v=0.4
)
```
